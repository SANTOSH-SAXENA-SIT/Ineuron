```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split , GridSearchCV
from sklearn.preprocessing import LabelEncoder
from xgboost import XGBClassifier
```

```
test = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.test" , skiprows=1 , header = None)
test.head()
```

```
train = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data' ,header = None)
train.head()
```

```
train.shape
```

```
test.shape
```

```
col_labels = ['age', 'workclass', 'fnlwgt', 'education', 'education_num',
'marital_status', 'occupation','relationship', 'race', 'sex', 'capital_gain',
'capital_loss', 'hours_per_week', 'native_country', 'wage_class']
```

```
train.columns = col_labels
test.columns = col_labels
```

```
train.head()
```

```
train.describe()
```

```
train[train.dtypes[train.dtypes == "object"].index]
```

```
ENCODER = []
for i in train[train.dtypes[train.dtypes == "object"].index].columns:
  encoder = LabelEncoder()
  train[i] = encoder.fit_transform(train[i])
  ENCODER.append(encoder)
```

```
train.head()
```

```
X = train.drop(["wage_class"] , axis = 1)
y = train[["wage_class"]]
```

```
X.head()
```

```
y.head()
```

```
x_train, x_test , y_train , y_test = train_test_split(X,y , test_size = 0.2 , random_state = 101)
```

```
model = XGBClassifier()
model.fit(x_train , y_train)
model.score(x_test , y_test)
```

```
param_grid={
   
    ' learning_rate':[1,0.5,0.1,0.01,0.001],
    'max_depth': [3,20],
    'n_estimators':[10,50]
    
}
```

```
grid= GridSearchCV(XGBClassifier(objective='binary:logistic'),param_grid, verbose=3)
```

```
grid.fit(x_train , y_train)
```

```
grid.best_params_
```

```
new_model = XGBClassifier(learning_rate= 1 , max_depth= 20 , n_estimators= 50)
new_model.fit(x_train , y_train)
new_model.score(x_test , y_test)
```
